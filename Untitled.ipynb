```python
def calcEbsSpeed(x):
    if x is np.nan:
        return np.nan
    if x.lower().startswith('upto '):
        x = x[5:]
    if x.lower().endswith(' mbps'):
        x = x[:-5]
    return float(x)
```

```python
def ebsSpeedIsUpTo(x):
    if x is np.nan:
        return False
    return x.lower().startswith('upto ')
```

```python
offers['ebsMbps'] = offers['ebsThroughput'].apply(lambda x: calcEbsSpeed(x))
```

```python
offers['ebsIsUpTo'] = offers['ebsThroughput'].apply(ebsSpeedIsUpTo)
```

```python
def calcNetSpeed(x):
    if x.lower().startswith('up to '):
        x = x[6:]
    if x.lower().endswith(' gigabit'):
        x = x[:-8]
    if x == "High":
        x = "1"
    if x == "Moderate":
        x = "0.3"
    return float(x)
```

```python
def netSpeedIsUpTo(x):
    return x.lower().startswith('up to')
```

```python
offers['netGbps'] = offers['networkPerformance'].apply(lambda x: calcNetSpeed(x))
```

```python
offers['netIsUpTo'] = offers['networkPerformance'].apply(lambda x: netSpeedIsUpTo(x))
```

```python
offers['memory'] = offers['memory'].apply(lambda x: float(x[:-4].replace(',', '')))
```

```python
offers['ecu'] = offers['ecu'].astype(float)
```

```python
offers['vcpu'] = offers['vcpu'].astype(int)
```

```python
offers[:10]
```

```python
offers.sort_values(by=['instanceFamily', 'OS', 'ecu'])[::15]
```

```python
final_offers = offers[['instanceType', 'instanceFamily', 'OS', 'ecu', 'vcpu', 'memory', 'normSizeFactor', 'storage', 'physicalProcessor', 'ebsMbps', 'ebsIsUpTo', 'netGbps', 'netIsUpTo']]
```

```python
cpuVcpuScale = 2
```

```python
source = final_servers[['numCpus', 'totalRamGB']].copy()
source['vcpuEstimate'] = source['numCpus'] * cpuVcpuScale
source.reset_index(inplace=True)
source[:5]
```

```python
temp = source[source['hostname']=='CTXUSSECDSK06PD']
temp
```

```python
def selectOffering(r):
        fit = final_offers[(final_offers['vcpu'] >= r['vcpuEstimate']) & (final_offers['instanceFamily'] != 'GPU instance') & (final_offers['memory'] > r.totalRamGB)]
        minVcpu = fit['vcpu'].min()
        fit2 = fit[fit['vcpu']==minVcpu].copy()
        fit2['hostname'] = r.hostname
        items.append(fit2)
```

```python
items = []
r = source.apply(selectOffering, axis=1)
matches = pd.concat(items)
matches
```
